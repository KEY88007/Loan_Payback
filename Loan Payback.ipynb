```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_train = df_train.drop('id' , axis=1)
df_test = df_test.drop('id' , axis=1)
```

```python
df_train['annual_income'].sort_values()
```

```python
set(df_test['education_level'].unique()).difference(set(df_train['education_level'].unique()))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
for i in df_test.columns:
    if df_test[i].dtype == object:
        df_train[i] = label.fit_transform(df_train[i])
        df_test[i] = label.transform(df_test[i])
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
X_train = df_train.drop('loan_paid_back',axis = 1)
y_train = df_train['loan_paid_back']
X_test = df_test.copy()
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train)
```

```python
model = Sequential()
model.add (Dense(units= 23, activation='relu', use_bias=False,input_shape = (11 , )))
model.add (Dense(units= 11, activation='relu', use_bias=False))
model.add (Dense(units= 2, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train , y_train, epochs=30, validation_split=0.16)
```

```python
X_train.shape
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = y_pred[::,1]
```

```python
df_sub['loan_paid_back'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index=False, index_label=False)
```
